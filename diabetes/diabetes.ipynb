# Descripcion del problema  

Descripcion de los atributos del dataset "Pima Indios Diabetes"

1. Número de veces embarazada
2. Concentración de glucosa en plasma a 2 horas en una prueba de tolerancia oral a la glucosa
3. Presión arterial diastólica (mm Hg)
4. Grosor del pliegue de la piel del tríceps (mm)
5. Insulina en suero de 2 horas (mu U / ml)
6. Índice de masa corporal (peso en kg / (altura en m) ^ 2)
7. Función de pedigrí de la diabetes
8. Edad (años)
9. Variable de clase (0 o 1)

# Modelo predictivo

```python
import sklearn
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn import metrics 
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

y = pd.DataFrame(data=df['target'])

X = pd.DataFrame(data=df.iloc[:,0:8])

X.head()
```

```python
y.head()
```

```python
from random import randint

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state=randint(0, 10000),stratify=y)

print (X_train.shape, y_train.shape) 
print (X_test.shape, y_test.shape)
```

```python
y_test.groupby('target')['target'].count()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf =  DecisionTreeClassifier ()
clf.fit(X_train, y_train)
C = clf.feature_importances_
print('Ponderacion de los atributos con sus respectiva correlacion con el target \n')
for i in range(len((C))):
    print(X.columns[i], ': ', round(C[i]*100,2))
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=0.1)
model_svm = svm.fit(X_train,y_train)
predictions_svm = model_svm.predict(X_test)

model_svm.score(X_test, y_test)
```

```python
print(classification_report(y_test, predictions_svm))
```

```python
predictions_svm
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, predictions_svm)

confusion_matrix
```

```python
plt.figure(figsize=(15, 8), dpi=70)
plt.subplot(1,1,1)

sns.heatmap(confusion_matrix, square = True, annot = True)
```

```python
from sklearn.linear_model import LogisticRegression

lm = LogisticRegression()
model_lm = lm.fit(X_train, y_train)
predictions = model_lm.predict(X_test)

model_lm.score(X_test, y_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, predictions)

confusion_matrix
```

```python
plt.figure(figsize=(15, 8), dpi=70)
plt.subplot(1,1,1)

sns.heatmap(confusion_matrix, square = True, annot = True)
```

```python
coincidencia = 0
for i in range(len(predictions_svm)):
    if predictions_svm[i] == predictions[i]:
        coincidencia = coincidencia + 1

coincidencia = (coincidencia / len(predictions_svm))*100

print("Coincidencia en los casos, comparando ambos modelos : ",round(coincidencia,2),"%")
```

# Curva ROC

```python
fig = plt.figure(figsize=(15,4))

fpr, tpr, thresholds = metrics.roc_curve(y_test, predictions_svm)
plt.plot(fpr, tpr, color = 'r', label = 'SVM')

fpr, tpr, thresholds = metrics.roc_curve(y_test, predictions)
plt.plot(fpr, tpr, color = 'b', label = 'LR')

plt.plot([0.0, 1.0],[0.0, 1.0],'g')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```
