# Descripción del dataset  
  
  
Titulo del estudio "Sitios de localización de proteínas en celulas"

Un sistema de clasificación probablistica utilizando Support Vector Machine para predecir sitios de localización de proteínas en las células.

-.Información del los atributos:

1-. **Nombre de secuencia**: Número de acceso para la base de datos SWISS-PROT  

2-. **mcg**: Método de McGeoch para reconocimiento de secuencia de señal.  

3-. **gvh**: Método de von Heijne para reconocimiento de secuencia de señal.  

4-. **lip**: Puntuación de la secuencia consenso de Signal Peptidase II de von Heijne. Atributo binario.  

5-. **chg**: Presencia de carga en el extremo N de las lipoproteínas predichas. Atributo binario.  

6-. **aac**: Puntuación del análisis discriminante del contenido de aminoácidos de membrana externa y proteínas periplásmicas.  

7-. **alm1**: Puntuación del programa de predicción de la región de membrana ALOM.  

8-. **alm2**: Puntuación del programa ALOM después de excluir la señal divisible putativa regiones de la secuencia.  
  
  
-.Distribución de clase. La clase es el sitio de localización. Por favor mira Nakai y Kanehisa mencionado anteriormente para más detalles.

1-. **cp**: Citoplasma  

2-. **im**: Membrana interna sin secuencia de señal 

3-. **pp**: Perisplasma 

4-. **imU**: Membrana interna, secuencia de señal no borrable

5-. **om**: Membrana externa

6-. **omL**: Lipoproteína de membrana externa 

7-. **imL**: Lipoproteína de membrana interna

8-. **imS**: Membrana interna, secuencia de señal divisible


Referencia: Paul Horton y Kenta Nakai, Sistemas Inteligentes en Biología Molecular, 109-115.

Mas información en https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.names

# Recopilación de datos  
  
Se hizo una  extracción de datos de la pagina UCI Machine Learning Repository, por lo que el set de datos viene listo para usarlo como data frame. El dataset se encuentra disponible en la siguiente dirección https://archive.ics.uci.edu/ml/datasets/Ecoli.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data",header=None, sep=r"\s*",names = ['sequence_name', 'mcg', 'gvh', 'lip','chg','aac','alm1', 'alm2','class_distribution'])

print("Total datos: ", len(df))

df.head()
```

# Modelo predictivo

En esta fase se aplicara un modelo predictivo en este caso particular se aplicara *Support Vector Machine* (**SVM**) utilizando la herramienta *scikit-learn* en python 3.6. Antes de aplicar el modelo, se deben preparar las variables dependiente e independiente asignando los *features* y *target* respectivamente. Luego de tener asignadas las variables, se realizara un split utilizando *train_test_split* utilizando estratificación.  

Después de aplicar el modelo **SVM** al dataset *train*, utilizamos la métrica de la curva **ROC** para estudiar la certeza del modelo utilizando los valores conocidos en el dataset *test*, en este caso el modelo da un *accuracy* del 98.5%. Podemos ademas estudiar el grado de certeza utilizando la matriz de confusión.

```python
import sklearn
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn import metrics 
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

y = pd.DataFrame(data=df['class_distribution'])

X = pd.DataFrame(data=df_fact.ix[:,0:8])

X.head()
```

```python
y.head()
```

```python
from random import randint

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state=randint(0, 10000),stratify=y)

print (X_train.shape, y_train.shape) 
print (X_test.shape, y_test.shape)
```

```python
y_test.groupby('class_distribution')['class_distribution'].count()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf =  DecisionTreeClassifier ()
clf.fit(X_train, y_train)
C = clf.feature_importances_
print('Ponderacion de los atributos con sus respectiva correlacion con el target \n')
for i in range(len((C))):
    print(X.columns[i], ': ', round(C[i]*100,2))
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=0.1)
model_svm = svm.fit(X_train,y_train)
predictions_svm = model_svm.predict(X_test)

model_svm.score(X_test, y_test)
```

```python
predictions_svm
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, predictions_svm)

confusion_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(15, 8), dpi=70)
plt.subplot(1,1,1)

sns.heatmap(confusion_matrix, square = True, annot = True)
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(1,2,1) 
ax.scatter(X_test['sequence_name'],predictions_svm,alpha=.5)
plt.grid(True, linewidth=0.15)
plt.title('Prediccion de correlacion entre nombre y clase')
plt.xlabel('sequence_name')
plt.ylabel('class_distribution')

ax = fig.add_subplot(1,2,2) 
ax.scatter(X_test['gvh'],predictions_svm,alpha=.5)
plt.grid(True, linewidth=0.15)
plt.title('Prediccion de correlacion entre gvh y clase')
plt.xlabel('gvh')
plt.ylabel('class_distribution')

fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(1,2,1) 
ax.scatter(X_test['mcg'],predictions_svm,alpha=.5)
plt.grid(True, linewidth=0.15)
plt.title('Prediccion de correlacion entre mcg y clase')
plt.xlabel('mcg')
plt.ylabel('class_distribution')

ax = fig.add_subplot(1,2,2) 
ax.scatter(X_test['alm1'],X_test['alm2'],alpha=.5)
plt.grid(True, linewidth=0.15)
plt.title('Prediccion de correlacion entre alm1 y alm2')
plt.xlabel('alm1')
plt.ylabel('alm2')

plt.show()
```
